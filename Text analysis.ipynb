```python
import re
import numpy as np
from scipy.spatial import distance
```

```python
file_obj = open('cats.txt','r')
```

```python
data_list = list(file_obj)
```

```python
data_list
```

```python
data_list2 = []
data_list3 = []
line_quantity = 0
```

```python
for line in data_list:
    data_list2.append(line.lower())
    line_quantity += 1
```

```python
for line in data_list2:
    line = re.split('[^a-z]', line)
    while '' in line:
        line.remove('')    
    data_list3.append(line)
```

```python
dict1 = {}
i = 0
```

```python
for line in data_list3:
    for word in line:
        if dict1.get(word) == None:
            dict1[word] = i
            i+=1
```

```python
dict1
```

```python
dict2 = {}
for i in range(0,len(dict1)):
    a,b = dict1.popitem()
    dict2[b] = a
dict2
```

```python
len(dict2)
```

```python
#matrix = np.ones((line_quantity,len(dict2)),int)
matrix = np.zeros( (line_quantity,len(dict2)) )
```

```python
matrix
```

```python
for i in range(0,line_quantity):
    for j in range(0,len(dict2)):
        if dict2[j] in data_list3[i]:
            matrix[i,j] += 1
```

```python
LenCosMin = 1
index_LenCosMin = 0
```

```python
for i in range(1, len(matrix) - 1):
    if distance.cosine(matrix[0], matrix[i]) < LenCosMin:
        LenCosMin = distance.cosine(matrix[0], matrix[i])
        index_LenCosMin = i
```

```python
print(LenCosMin)
print(index_LenCosMin)
```
